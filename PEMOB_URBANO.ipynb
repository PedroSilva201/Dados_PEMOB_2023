```python
import pandas as pd

arquivo = "pemob2023_municipal.xlsx__pemob_municipal_2023.csv"
df = pd.read_csv(arquivo)

```

```python
print(df.columns)  # mostra todos os nomes de colunas
```

```python
# Get the list of all columns
# Apresenta sobre os dados correspondentes a coluna Y dentro deste questionário Qual o número de ônibus convencionais operacionais de piso baixo no município?
all_columns = df.columns.tolist()

# Determine the index for column 'Y' (25th column, 0-indexed is 24)
# Make sure the index is within the bounds of the column list
column_y_index = 24

if column_y_index < len(all_columns):
    column_y_name = all_columns[column_y_index]

    print(f"Processing column: '{column_y_name}' (corresponds to column Y)")

    # Ensure the column is numeric, converting errors to NaN and filling NaN with 0
    df[column_y_name] = pd.to_numeric(df[column_y_name], errors="coerce").fillna(0)

    # Check the data type of the column after processing
    print(f"Data type of column '{column_y_name}' after processing: {df[column_y_name].dtype}")
    # Print unique values to see what's left (should ideally be only numbers and 0)
    print(f"Unique values in column '{column_y_name}' after processing: {df[column_y_name].unique()}")


    # Calculate the sum of the specified column
    soma_coluna_y = df[column_y_name].sum()

    print(f"\nA soma total da coluna '{column_y_name}' (coluna Y) é: {soma_coluna_y}")
else:
    print(f"Não há uma coluna na posição 'Y' (índice {column_y_index}) no DataFrame. O DataFrame tem apenas {len(all_columns)} colunas.")
```

```python
# Aborda o exemplo da coluna X sobre onibus que possuem plataforma elevatória
# Plataforma elevatória é um degrau movel que pode ser trasnformado em elevador
all_columns = df.columns.tolist()


column_x_index = 25

if column_x_index < len(all_columns):
    column_x_name = all_columns[column_x_index]

    print(f"Processing column: '{column_x_name}' (corresponds to column X)")

    # Ensure the column is numeric, converting errors to NaN and filling NaN with 0
    df[column_x_name] = pd.to_numeric(df[column_x_name], errors="coerce").fillna(0)

    # Check the data type of the column after processing
    print(f"Data type of column '{column_x_name}' after processing: {df[column_x_name].dtype}")
    # Print unique values to see what's left (should ideally be only numbers and 0)
    print(f"Unique values in column '{column_x_name}' after processing: {df[column_x_name].unique()}")


    # Calculate the sum of the specified column
    soma_coluna_x = df[column_x_name].sum()

    print(f"\nA soma total da coluna '{column_x_name}' (coluna X) é: {soma_coluna_x}")
else:
    print(f"Não há uma coluna na posição 'X' (índice {column_x_index}) no DataFrame. O DataFrame tem apenas {len(all_columns)} colunas.")
```
